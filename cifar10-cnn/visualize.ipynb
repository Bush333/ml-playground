```python
from __future__ import print_function
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from keras.datasets import cifar10
```

```python
# Load CIFAR10, and split into training and test sets
(X_train, y_train), (X_test, y_test) = cifar10.load_data()
print('X_train shape: ', X_train.shape)
print(X_train.shape[0], 'training samples')
print(X_test.shape[0], 'test samples')
```

```python
categories = []
categories.append('airplane')
categories.append('automobile')
categories.append('bird')
categories.append('cat')
categories.append('deer')
categories.append('dog')
categories.append('frog')
categories.append('horse')
categories.append('ship')
categories.append('truck')

# Show first 10 images from each of the categories
fig = plt.figure()
gs = gridspec.GridSpec(10, 10)

count = 0
gs_idx = 0

for j in range(len(categories)):
    for i in range(X_train.shape[0]):
        if (int(y_train[i]) == j and count < 10):
            sample_img = X_train[i, :, :, :]
            fig.add_subplot(gs[gs_idx])
            plt.imshow(sample_img)
            plt.axis('off')
            gs_idx += 1
            count += 1
    count = 0
plt.show()
```
